## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#access gender column and separate male and female
all_players=purchase_data.loc[purchase_data["Gender"] =="Male"]
male_players = len(all_players)
male_players


all_players=purchase_data.loc[purchase_data["Gender"] =="Female"]
female_players = len(all_players)
female_players


all_players =purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed"]
other_players = len(all_players)
other_players

#determine % of each gender
male_player_perc = male_players / total_players * 100
male_player_perc

female_player_perc = female_players / total_players * 100
female_player_perc

other_player_perc = other_players / total_players * 100
other_player_perc

#display in dataframe
gender_data = pd.DataFrame({
        "Percentage of Players":[male_player_perc,female_player_perc,other_player_perc],
    "Total Count":[male_players,female_players,other_players],
                  "Category":["Male", "Female", "Other/Non-Disclosed"]
}).round(2)
gender_data_arrange =gender_data.set_index("Category")
gender_data_arrange.head()
```

```python
#assign/organize with age bins
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age_binning"] = pd.cut(purchase_data["Age"],age_bins, labels = group_names)
purchase_data
purchase_data.head()
```

```python
# Create Groupby Object
age_group = purchase_data.groupby("Age_binning")
#determine rows for each bin
totplayercalc = age_group["Age"].count()
# Calculation for % of players
percnt = totplayercalc / 576 * 100
age_demo_df = pd.DataFrame({"Percentage of Players" :percnt,
                          "Total Count" :totplayercalc
                                 }).round(2)
age_demo_df.head()
```

```python
age_demos1 = age_demo_df.reindex(columns=["Percentage of Players","Total Count"])
age_demos1


purchase_data["Age_binning"] = pd.cut(purchase_data["Age"],age_bins, labels = group_names)
agedemo2 = purchase_data.groupby("Age_binning")
purct = agedemo2["SN"].count()
avgpurpri = agedemo2["Price"].mean()
totpurval = agedemo2["Price"].sum()
avgpurtotperper= totpurval/purct

newagedf = pd.DataFrame({"Purchase Count":purct,
                        "Average Purchase Price":avgpurpri,
                        "Total Purchase Value":totpurval,
                        "Average Purchase Total Per Person":avgpurtotperper}).round(2)
newagedf1 = newagedf.reindex(columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Purchase Total Per Person"])
newagedf1.style.format({"Average Purchase Price":"${:.2f}", "Total Purchase Value":"${:.2f}","Avg Purchase Total Per Person":"${:.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#perform mean/count/sum on the groups
player_grp = purchase_data.groupby("SN")
itemidct = player_grp["Price"].count()
avgpp= player_grp["Price"].mean()
totpv = player_grp["Price"].sum()

#display in a dataframe
newspenderdf = pd.DataFrame({
                            "Purchase Count" :itemidct,
                            "Average Purchase Price" :avgpp,
                            "Total Purchase Value" : totpv}).round(2)
newspenderdf=newspenderdf.sort_values(by ="Total Purchase Value", ascending = False)
newspenderdf = newspenderdf.reindex(columns=["Purchase Count","Average Purchase Price","Total Purchase Value"])
newspenderdf.style.format({"Average Purchase Price":"${:.2f}","Total Purchase Value": "${:.2f}"})
newspenderdf.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

player_grp2 = purchase_data.groupby(["Item ID","Item Name"])

itemid1 = player_grp2["Item Name"].unique
itemidct1 = player_grp2["Price"].count()
avgpp1= player_grp2["Price"].unique()
totpv1 = player_grp2["Price"].sum()


popularitems = pd.DataFrame({"Purchase Count" :itemidct1,
                            "Item Price" :avgpp1,
                            "Total Purchase Value" : totpv1}).round(2)
popularitems=popularitems.sort_values(["Purchase Count","Total Purchase Value"], ascending = False)
popularitems = popularitems.reindex(columns=["Purchase Count","Item Price","Total Purchase Value"])
popularitems.head()
```
